#With no hidden layer

```
import numpy as np

weight = np.array([0.7,0.2,-0.5,])
alpha = 0.1
```

```
streetLight = np.array([[0,0,1],
                        [0,1,1],
                        [0,0,1],
                        [1,1,1],
                        [0,1,1],
                        [1,0,1]])

# Stop = 0
# Walk = 1

walkStop = np.array([0,1,0,1,1,0])
```

```
print(len(streetLight))
```

```
for epoch in range(40):
  total_error = 0
  
  print("Epoch :", epoch+1)
  print("----------------------------------")
  print(walkStop)
  print("[", end = " ")

  for eachRow in range(len(walkStop)):
    input = streetLight[eachRow]
    target = walkStop[eachRow]

    # simple one layer prediction algo
    prediction = input.dot(weight)
    error = (target - prediction) ** 2
    total_error +=error
    delta = prediction - target 
    weight = weight - (alpha * input * delta)
    #print("Prediction for " + str(eachRow+1) + " : " + str(round(prediction,2)),end = ", "),
    print(round(prediction,2), end = ", ")

  print("]")

  print("Weight : " , weight)
  print("Error : ", error )

  print("----------------------------------")
```

## Error improved from 0.305 to 0.00000344
